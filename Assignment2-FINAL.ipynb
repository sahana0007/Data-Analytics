```python
from tensorflow.keras.optimizers import Adam
from keras_tuner.tuners import Hyperband, RandomSearch
from keras.utils.generic_utils import get_custom_objects
from tensorflow.keras.layers import Dense, Dropout, LeakyReLU, Activation, LSTM, Flatten, SimpleRNN
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
get_custom_objects().update({'leaky_relu': Activation(LeakyReLU(alpha=0.2))})
def build_model(hp):
    model = Sequential()
    model.add(Dense(units=hp.Int('units', min_value=8, max_value=512, step=32),
                    activation=hp.Choice('act', ['relu'])))
    for i in range(hp.Int('layers', 2, 4)):
        model.add(Dense(units=hp.Int('units_' + str(i), min_value=8, max_value=512, step=32),
                        activation=hp.Choice('act_' + str(i), ['relu'])))
        model.add(Dropout(hp.Choice('drop_out', [0.2, 0.3, 0.4])))

    model.add(Dense(1, activation='linear'))
    model.compile(optimizer=Adam(hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])), loss='mean_squared_error',
                  metrics='mean_squared_error')
    return model
tuner = RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=3,
    executions_per_trial=2,
    overwrite=True,
    directory='project',
    project_name='Resale Price prediction'
    )
print(tuner.search_space_summary())

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

tuner.search(X_train, y_train, epochs=50, callbacks=[early_stopping]
             , batch_size=256, shuffle=False, verbose=1, validation_data=(X_test, y_test))

print(tuner.results_summary())

# Get best hyperparameters
optimal_hp = tuner.get_best_hyperparameters(num_trials=1)[0]
# print(best_hps.get('units_0'), best_hps.get('learning_rate'), best_hps.get('layers'), best_hps.get('act0'))


# build the FF model with optimal hyper-parameters
model = tuner.hypermodel.build(optimal_hp)
```

```python
# fit the best model on the dataset
#model.fit(X_train, y_train, epochs=70, batch_size=256, verbose=1)

# fit the best model on the dataset
history = model.fit(X_train, y_train, epochs=50, batch_size=256, verbose=1, callbacks=[early_stopping])
loss_per_epoch = history.history['loss']
#loss_per_epoch_df = pd.DataFrame({"loss": loss_per_epoch, "loss": history.history['val_loss']})
plt.plot(range(len(loss_per_epoch)), loss_per_epoch)
```
